# 15: 금액의 결측치 수정

### 데이터에 결측치가 있는지 확인
- isnull() : 데이터의 결측치 유무 확인

```python
uriage_data.isnull().any(axis=0)
```

- any(): 하나라도 값이 True가 있으면 True를 반환하는 함수
- axis=0: 각 열을 기준으로 검사

### 데이터의 결손치 수정

```python
flg_is_null = uriage_data["item_price"].isnull()
flg_is_null
```

```python
for trg in list(uriage_data.loc[flg_is_null, "item_name"].unique()):
    price = uriage_data.loc[(~flg_is_null) & (uriage_data["item_name"] ==trg), "item_price"].max()
    uriage_data["item_price"].loc[(flg_is_null) & (uriage_data["item_name"]==trg)] = price
```

위의 코드가 오류 뜨는 관계로 수정된 코드 진행
- uriage_data["item_price"].loc[]을 사용해 체이닝 방식으로 데이터를 수정하려다 보니, Pandas가 복사본인지 원본인지 혼동하게 되어 경고 발셍
- loc[]을 이용해 한 번에 DataFrame을 직접 수정하도록 코드를 변경

```python
flg_is_null = uriage_data["item_price"].isnull() #결측치 있는 곳 조사

for trg in list(uriage_data.loc[flg_is_null, "item_name"].unique()):
    price = uriage_data.loc[(~flg_is_null) & (uriage_data["item_name"] == trg), "item_price"].max()
    uriage_data.loc[(flg_is_null) & (uriage_data["item_name"] == trg), "item_price"] = price
```

[코드 분석]

# 17: 날짜 오류 수정
- 엑셀 데이터 다룰 때 주의할 점: 서식이 다른 데이터가 섞여 있을 수 있음

```python
flg_is_serial = kokyaku_data['등록일'].astype("str").str.isdigit() #고객 정보의 등록일이 숫자인지 아닌지
flg_is_serial.sum() #추후 검증을 위해 개수 표시
```

- astype(): 원하는 데이터타입으로 바꿈
- str.isdigit(): 숫자인지 아닌지 판정

### 숫자로 등록된 등록일 부분 수정

```python
fromSerial = pd.to_timedelta(kokyaku_data.loc[flg_is_serial, '등록일'].astype("float"), unit="D") + pd.to_datetime("1900/01/01")
fromSerial
```

- 엑셀에서는 날짜를 'Serial Number' 형식으로 저장한다
    - ex. 1900/01/01 -> 1, 2020/12/31 -> 44197 로 저장
- 엑셀의 이러한 Serial Number를 실제 날짜로 변환시켜야 한다.
- 따라서 1900년 01월 01일을 기준으로 날짜 차이를 더하여 최종 등록일을 구한다
- pd.to_timedelta(): 숫자 -> 날짜

### 날짜로 변환된 데이터도 서식 통일
- '/'로 구분된 서식 -> '-'으로 구분된 서식

```python
fromString = pd.to_datetime(kokyaku_data.loc[~flg_is_serial, '등록일'])
fromString
```

```python
kokyaku_data['등록일'] = pd.concat([fromSerial, fromString])
kokyaku_data
```

```python
kokyaku_data['등록연월'] = kokyaku_data['등록일'].dt.strftime("%Y%m")
rslt = kokyaku_data.groupby('등록연월').count()['고객이름']
print(rslt)
print(len(kokyaku_data))
```

### 등록일 칼럼에 숫자 데이터 남아있는지 확인

```python
flg_is_serial = kokyaku_data['등록일'].astype("str").str.isdigit()
flg_is_serial.sum()
```

- pd.concat(): 여러 DataFrame이나 Series를 수직 또는 수평으로 결합하여 하나로 만드는 함수
- fromSerial과 fromString 두 개의 Series를 하나로 합쳐서 등록일이라는 하나의 열을 만듦

# 18: 두 개의 데이터 결합

### 매출 이력, 고객 정보 결합

```python
join_data = pd.merge(uriage_data, kokyaku_data, left_on="customer_name", right_on="고객이름", how="left")
join_data = join_data.drop("customer_name", axis=1)
join_data
```

- pd.merge()
    - left_on: 왼쪽 DataFrame에서 결합 기준으로 사용할 열 지정
    - right_on: 오른쪽 DataFrame에서 결합 기준으로 사용할 열 지정
    - how="left": left join 방식(uriage_data의 모든 데이터 유지, kokyaku_dzata에서 일치하는 데이터만 결합)

- drop()
    - 특정 열이나 행을 제거하는 함수
    - axis=1: 열을 기준으로 제거
    

    
- 데이터 결합 과정에서 'customer_name'과 '고객이름'이라는 두 열이 동일한 정보를 담고 있기 때문에 하나를 drop함
- 여기까지 했으면 '데이터 정제' 완료
